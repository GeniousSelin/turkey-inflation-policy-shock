# Comparative Analysis: Global Pandemic Shock vs. Unorthodox Monetary Policy ("Erdoganomics") 
**The focus of my project:** Quantitative Macroeconomics & Financial Risk Assessment

International corporations face two types of financial crises when operating in emerging markets:
1. **External Shocks:** Macroeconomic disruptions affecting the global economy simultaneously (e.g., the COVID-19 pandemic).
2. **Policy Shocks:** Internal disruptions caused by domestic government interventions (e.g., Turkey's decision to cut interest rates during skyrocketing inflation).

I wanted to test the claim made by economists that internal policy is sometimes more dangerous to a company's bottom line than a literal global pandemic. I used Turkey's hyperinflation landscape as a stress test. Instead of using external dependencies like Pandas, I chose to demonstrate strong native Python fundamentals. I handled explicit mapping using dictionary items, list filtering conditions, and manually coded the statistical metrics.

# Step 1 Creating Database 
I am building my own database using a **Python dictionary**. The **strings represent the dates**, and **the values are floats representing the official annual inflation rate** for that month.

The COVID-19 period: from March 2020 to the end of 2021 (when the storm raged across the globe).
The Erdoganomics period: from the end of 2021 to 2024 (when aggressive rate cuts began, contrary to market logic).

```python
turkey_inflation_db = {
    "2020-03": 11.86, "2020-06": 12.62, "2020-09": 11.75, "2020-12": 14.60,
    "2021-03": 16.19, "2021-06": 17.53, "2021-09": 19.58, "2021-11": 21.31,
    
    "2021-12": 36.08, 
    "2022-03": 61.14, "2022-06": 78.62, "2022-09": 83.45, "2022-12": 64.27,
    "2023-03": 50.51, "2023-06": 38.21, "2023-09": 61.53, "2023-12": 64.77,
    "2024-03": 68.50, "2024-06": 71.60
}
print(f"Database successfully initialized with {len(turkey_inflation_db)} data points.")
```

# Step 2 Data Segregation

```python
covid_period_data = []
policy_shock_data = []

for date, inflation_rate in turkey_inflation_db.items():
    if "2020" in date or "2021-03" in date or "2021-06" in date or "2021-09" in date or "2021-11" in date:
        covid_period_data.append(inflation_rate)
    else:
        policy_shock_data.append(inflation_rate)

print(f"Pandemic Era Dataset: {len(covid_period_data)} quarterly checkpoints extracted.")
print(f"Policy Shock Dataset: {len(policy_shock_data)} quarterly checkpoints extracted.")
```

# Step 3 Financial Metrics 

```python
def compute_macro_metrics(inflation_list):
    if len(inflation_list) == 0:
        return 0, 0, 0
    
    average_rate = sum(inflation_list) / len(inflation_list)
    
    
    peak_rate = max(inflation_list)
    
   
    total_acceleration = inflation_list[-1] - inflation_list[0]
    
    return average_rate, peak_rate, total_acceleration

covid_avg, covid_peak, covid_acceleration = compute_macro_metrics(covid_period_data)
policy_avg, policy_peak, policy_acceleration = compute_macro_metrics(policy_shock_data)

print("[PHASE 1: COVID-19 PANDEMIC GLOBAL ERA]")
print(f"-> Average Baseline Inflation: {covid_avg:.2f}%")
print(f"-> Worst-Case Peak Inflation:  {covid_peak:.2f}%")
print(f"-> Inflation Growth Velocity: +{covid_acceleration:.2f}%")

print("[PHASE 2: ERDOGANOMICS INTERNAL POLICY ERA]")
print(f"-> Average Baseline Inflation: {policy_avg:.2f}%")
print(f"-> Worst-Case Peak Inflation:  {policy_peak:.2f}%")
print(f"-> Inflation Growth Velocity: +{policy_acceleration:.2f}%")
```

# Step 4 Analytical Conclusion

```python
if policy_peak > covid_peak:
    severity_multiplier = policy_peak / covid_peak
    print("Conclusion:")
    print("The internal political experiment ('Erdoganomics' unconventional interest rate cuts was exponentially more toxic to corporate stability than the COVID-19 pandemic.")
    print(f"Peak systemic inflation reached during the policy shock was {severity_multiplier:.1f}x higher than during the entire global healthcare emergency crisis.")
    print("\nSTRATEGIC RISK ADVICE FOR THE BOARD:")
    print("Avoid long-term capital exposure in markets practicing non-traditional economic theories.")
else:
    print("Conclusion:")
    print("Global systemic market shocks outweighed domestic micro-policy decisions.")
```
